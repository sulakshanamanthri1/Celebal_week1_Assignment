```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/pc/Desktop/Week2_Celebal/data/raw/tesla_deliveries_dataset_2015_2025.csv")

# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Distribution of Avg Price
sns.histplot(df["Avg_Price_USD"], bins=30, kde=True)
plt.title("Distribution of Tesla Average Price")
plt.show()
```
